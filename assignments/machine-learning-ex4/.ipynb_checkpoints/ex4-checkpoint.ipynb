```python
def getDatumImg(row):
    """
    Function that is handed a single np array with shape 1x400,
    crates an image object from it, and returns it
    """
    width, height = 20, 20
    square = row[1:].reshape(width,height)
    return square.T
    
def displayData(indices_to_display = None):
    """
    Function that picks 100 random rows from X, creates a 20x20 image from each,
    then stitches them together into a 10x10 grid of images, and shows it.
    """
    width, height = 20, 20
    nrows, ncols = 10, 10
    if not indices_to_display:
        indices_to_display = random.sample(range(X.shape[0]), nrows*ncols) # randomly select 100 pictures to display
        
    big_picture = np.zeros((height*nrows,width*ncols))
    
    irow, icol = 0, 0
    for idx in indices_to_display:
        if icol == ncols:
            irow += 1
            icol  = 0
        iimg = getDatumImg(X[idx])
        big_picture[irow*height:irow*height+iimg.shape[0],icol*width:icol*width+iimg.shape[1]] = iimg
        icol += 1
    fig = plt.figure(figsize=(6,6))
    img = Image.fromarray(big_picture*255)
    plt.imshow(img,cmap = cm.Greys_r)
```

### 1.3 Feedforward and cost function

```python
def propagateForward(X, Thetas):
    """
    X is (m, nx) darray
    X Thetas is l length 2D array
    set activate function sigmod, no batch
    no bias unit, each propagation cencentrate a 1s column to x 
    return the last layer output
    """
    L = len(Thetas) # L layers
    Z = []
    A = []
    m = X.shape[0] #m examples
    output = X.copy() #backup input X
    for l in range(L):
        Theta = Thetas[l] #l th layer, 
        z = np.dot(output, Theta.T).reshape([m, Theta.shape[0]]) #z=wx
        a = sigmod(z) #activate
        Z.append(z)
        A.append(a)
        if(l == L - 1):
            return a, Z, A #last layer needn't bias column
        a = np.insert(a, 0, 1, axis = 1) #insert a bias column of 1s
        output = a
```

```python
Thetas = [theta1, theta2]
output, dummy1, dummy2 = propagateForward(X, Thetas)
output.shape
```

```python
#recode y to vector 
m = y.shape[0]
recodedY = np.zeros([m, 10]) #(5000,1) -> (5000,10)
for i in range(m):
    n = y[i][0] - 1
    recodedY[i][n] = 1
recodedY.shape
```

```python
def computeCost(Thetas, X, Y, lamda = 0):
    m = X.shape[0]
    h, dummy1, dummy2 = propagateForward(X, Thetas) #(m,ny)
    cost = -Y * np.log(h) - (1 - Y) * np.log(1 - h)
    cost = np.sum(cost.flatten())/float(m)
    reg = 0.
    for i in range(len(Thetas)):
        t = Thetas[i].flatten()
        reg += np.sum(t**2)
    reg = reg / (2.*m) * lamda
    J = cost + reg
    return J
```

## 2 Backpropagation

### 2.1 Sigmoid gradient

```python
#derivative of sigmod function
def sigmodGradient(z):
    g = sigmod(z)
    return g*(1 - g)
```

### 2.2 Random initialization

```python
def genRandThetas():
    epsilon_init = 0.12
    input_layer_size = 400
    hidden_layer_size = 25
    output_layer_size = 10 

    theta1_shape = (hidden_layer_size, input_layer_size+1)
    theta2_shape = (output_layer_size, hidden_layer_size+1)
    rand_thetas = [ np.random.rand( *theta1_shape ) * 2 * epsilon_init - epsilon_init, \
                    np.random.rand( *theta2_shape ) * 2 * epsilon_init - epsilon_init]
    return rand_thetas
```

```python
ThetasInit = genRandThetas()
[x.shape for x in ThetasInit]
```

### 2.3 Backpropagation

```python
def propagationBackward(Thetas, X, Y, lamda = 0.):
    L = len(Thetas) #length of layers
    m = X.shape[0] #exmaples number
    delta, D, Delta = [], [], []
    delta.append(np.zeros([1,Thetas[0].shape[1]-1]))#init one more 
    for l in range(L):
        dim1 = Thetas[l].shape[1] #include bias unit 
        dim2 = Thetas[l].shape[0] 
        Delta.append(np.zeros([dim2,dim1])) #init Delta
        D.append(np.zeros([dim2,dim1])) #init D
        delta.append(np.zeros([1,dim2]))
        
    for i in range(m): #stochastic gradient descent
        x = X[i].reshape([1,X.shape[1]])
        y = Y[i].reshape([1,Y.shape[1]])
        
        yPredict, z, a = propagateForward(x, Thetas)
        delta[L] = yPredict - y 
        
        for j in range(1,L):
            l = L - j 
            delta[l] = delta[l+1].dot(Thetas[l][:,1:]) * sigmodGradient(z[l-1]) #(1,25)
            a[l-1] = np.insert(a[l-1],0,1,axis=1)
            Delta[l] += np.dot(delta[l+1].T,a[l-1])     
        Delta[0] += np.dot(delta[1].T,x) 
    for l in range(L):
        D[l] = Delta[l]/float(m)
        D[l][:,1:] = D[l][:,1:] + (float(lamda)/float(m))*Thetas[l][:,1:]
    return D

D = propagationBackward(Thetas, X[:5,:], recodedY[:5,:])
[x.shape for x in D]
```

### 2.4 Gradient checking

```python
#flatten and reshape utils
n_training_samples = 5000
input_layer_size = 400
hidden_layer_size = 25
output_layer_size = 10 

def flattenParams(thetas_list):
    """
    Hand this function a list of theta matrices, and it will flatten it
    into one long (n,1) shaped numpy array
    """
    flattened_list = [ mytheta.flatten() for mytheta in thetas_list ]
    combined = list(itertools.chain.from_iterable(flattened_list))
    assert len(combined) == (input_layer_size+1)*hidden_layer_size + \
                            (hidden_layer_size+1)*output_layer_size
    return np.array(combined).reshape((len(combined),1))

def reshapeParams(flattened_array):
    theta1 = flattened_array[:(input_layer_size+1)*hidden_layer_size] \
            .reshape((hidden_layer_size,input_layer_size+1))
    theta2 = flattened_array[(input_layer_size+1)*hidden_layer_size:] \
            .reshape((output_layer_size,hidden_layer_size+1))
    return [ theta1, theta2 ]

def flattenX(X):
    return np.array(X.flatten()).reshape((n_training_samples*(input_layer_size+1),1))

def reshapeX(flattenedX):
    return np.array(flattenedX).reshape((n_training_samples,input_layer_size+1))
```

```python
def computeCostFlattened(ThetasFlattend, XFlattened,Y, lamda = 0.):
    Thetas = reshapeParams(ThetasFlattend)
    X = reshapeX(XFlattened)
    return computeCost(Thetas, X, Y, lamda)
    
def propagationBackwardFlattened(ThetasFlattened,XFlattened,Y,lamda=0.):
    Thetas = reshapeParams(ThetasFlattened)
    X = reshapeX(XFlattened)
    Ds = propagationBackward(Thetas, X, Y, lamda)
    return flattenParams(Ds).flatten()
```

```python
def gradientCheck(Thetas, D, X, Y, lamda = 0.):
    epsilon = 0.0001
    m = X.shape[0]
    nx = X.shape[1]
    ny = Y.shape[1]
    flattened = flattenParams(Thetas)
    flattenedD = flattenParams(D)
    flattenedX = flattenX(X)
    nElems = len(flattened)
    for i in range(10):
        idx = int(np.random.rand()*nElems) #random select from 1-m
        epsvec = np.zeros([nElems, 1])
        epsvec[idx] = epsilon
        c1 = computeCostFlattened(flattened + epsvec,flattenedX,Y,lamda)
        c2 = computeCostFlattened(flattened - epsvec,flattenedX,Y,lamda)
        grad = (c1 - c2) / float(2. * epsilon)
        print('%dth example, numerical grad = %lf, backprop grad = %lf'\
              %(idx, grad, flattenedD[idx]))
```

```python
Ds = propagationBackward(Thetas, X, recodedY)
gradientCheck(Thetas, Ds, X, recodedY)
```

### 2.5 Regularized Neural Networks(Done)

```python
#train using gradient descent
Thetas = ThetasInit
alpha = 0.01
iterations = 1000
costHistory = []
for iter in range(1,iterations):
    dDs = propagationBackward(Thetas, X, recodedY, lamda = 0.)
    Thetas = [Thetas[i]-dD for i,dD in enumerate(dDs)]
    cost = computeCost(Thetas,X,recodedY)
    costHistory.append(cost)

xvals = range(len(costHistory))
plt.figure(figsize = (10,7))
plt.title('cost - iterations')
plt.xlabel('iteration no.')
plt.ylabel('cost')
plt.plot(xvals,costHistory)
print(np.min(costHistory))
```

### 2.6 Learning parameters using fmincg

```python
def trainNN(lamda=0.):
    ThetasInit = flattenParams(genRandThetas())
    result = optimize.fmin_cg(computeCostFlattened,\
                            x0=ThetasInit,\
                            fprime=propagationBackwardFlattened,\
                            args=(flattenX(X),recodedY,lamda),\
                            maxiter=50,disp=True,full_output=True)
    return reshapeParams(result[0])
```

```python
OptimaThetas = trainNN()
```

```python
#return scalar number idx from 1-10
def predictNN(Xrow,Thetas):
    output, Z, A = propagateForward(Xrow, Thetas) 
    #output is probbility vector in (0,1)
    result = np.argmax(output.flatten())+1 #(0,9) -> (1,10)
    return result

#evaluate the model, compute accuracy
def computeAccuracy(X,Thetas,y):
    correctN = 0
    m = X.shape[0]
    for i in range(m):
        pnum = predictNN(X[i].reshape([1,X.shape[1]]), Thetas)
        if(pnum == y[i][0]):
            correctN = correctN + 1
    return correctN / X.shape[0]
```

```python
computeAccuracy(X,OptimaThetas,y)
```

## 3 Visualizing the hidden layer

```python
def displayHiddenLayer(myTheta):
    """
    Function that takes slices of the first Theta matrix (that goes from
    the input layer to the hidden layer), removes the bias unit, and reshapes
    it into a 20x20 image, and shows it
    """
    #remove bias unit:
    myTheta = myTheta[:,1:]
    assert myTheta.shape == (25,400)
    
    width, height = 20, 20
    nrows, ncols = 5, 5
        
    big_picture = np.zeros((height*nrows,width*ncols))
    
    irow, icol = 0, 0
    for row in myTheta:
        if icol == ncols:
            irow += 1
            icol  = 0
        iimg = getDatumImg(np.insert(row,0,1))
        big_picture[irow*height:irow*height+iimg.shape[0],icol*width:icol*width+iimg.shape[1]] = iimg
        icol += 1
    fig = plt.figure(figsize=(6,6))
    minv,maxv = np.min(big_picture),np.max(big_picture)
    big_picture = (big_picture - minv)/(maxv-minv)*255
    img = Image.fromarray( big_picture)
    plt.imshow(img,cmap = cm.Greys_r)
    
displayHiddenLayer(OptimaThetas[0])
```
