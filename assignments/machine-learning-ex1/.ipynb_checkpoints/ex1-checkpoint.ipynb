## Linear regression with one variable
In this part of this exercise, you will implement linear regression with one variable to predict profits for a food truck. Suppose you are the CEO of a restaurant franchise and are considering different cities for opening a new outlet. The chain already has trucks in various cities and you have data for profits and populations from the cities.

You would like to use this data to help you select which city to expand to next.

The file ex1data1.txt contains the dataset for our linear regression problem. The first column is the population of a city and the second column is he profit of a food truck in that city. A negative value for profit indicates a oss.

### 2.1 Plotting the Data

```python
import pandas as pd
from matplotlib import pyplot as plt
data = pd.read_csv('ex1/ex1data1.txt',header=None);
data
```

```python
plt.ylabel('Profit in $10,000s')
plt.xlabel('Population of City in 10,000s')
plt.plot(data[0],data[1],'rx')
plt.show()
```

### 2.2 Gradient Descent

```python
#set up the data
import numpy as np
X = np.c_[np.ones(len(data[0])),data[0]] #add a column of ones to x
y = np.reshape(data[1].values,(97,1),order='C')
m = len(X)
theta = np.zeros([2,1]) #initialize fitting parameters
iterations = 1500
alpha = 0.01 #learning rate

#hypothesis function 
def h(theta,X):
    return np.dot(X,theta)

#computing the cost J(θ)
def J(theta,X,y):
    m = len(X)
    cost = 0.5/m*np.dot((h(theta,X)-y).T,h(theta,X)-y)
    return cost[0][0]

print('cost = ',J(theta,X,y))
```

```python
#gradient descent
def gradientDescent(X,y,theta,alpha,iterations):
    m = len(X)
    thetaHistory = [theta] #for plotting theta-iter
    costHistory = [J(theta,X,y)] #for plotting J-iter
    #start iteration
    for iter in range(1,iterations):
        theta = theta - alpha*1/m*np.dot(X.T,h(theta,X)-y)
        thetaHistory.append(theta)
        costHistory.append(J(theta,X,y))
    return theta, thetaHistory, costHistory

theta, thetaHistory, costHistory = gradientDescent(X,y,theta,alpha,iterations)
print('by gradient descent, theta = ',theta)
```

### Normal Equation

```python
#use normal equation
def normalEquation(X,y):
    return np.dot(np.dot(np.linalg.inv(np.dot(X.T,X)),X.T),y)

print('by normal equation, theta = ',normalEquation(X,y))
```

### 2.3 Debugging

```python
def plotConvergence(costHistory):
    plt.figure(figsize=(10,6))
    plt.grid(True)
    plt.title("Convergence of Cost Function")
    plt.xlabel("Iteration number")
    plt.ylabel("Cost function")
    plt.xlim([-0.1*iterations,1.1*iterations])
    plt.ylim([4,7])
    plt.plot(range(len(costHistory)),costHistory,'bo')
    
plotConvergence(costHistory)
```

```python
plt.figure(figsize=(10,6))
plt.plot(data[0],data[1],'rx',label='Training Set')
plt.plot(X[:,1],h(theta,X),label='Hypothesis: h(x) = %0.2f + %0.2fx'%(theta[0],theta[1]))
plt.grid(True)
plt.ylabel('Profit in $10,000s')
plt.xlabel('Population of City in 10,000s')
plt.legend()
```

### 2.4 Visualizing J(θ)

```python
from mpl_toolkits.mplot3d import Axes3D

theta0 = np.arange(-10, 10, 0.5)
theta1 = np.arange(-1, 4, 0.1)
xvals,yvals = np.meshgrid(theta0,theta1)
xvals,yvals,zvals = [],[],[]
for v0 in theta0:
    for v1 in theta1:
        zvals.append(J(np.array([v0,v1]),X,y))
        xvals.append(v0)
        yvals.append(v1)
        
fig = plt.figure(figsize=(10,10))
ax = Axes3D(fig)     #generate 3D object        
ax.set_xlabel(r'$\theta_0$',fontsize=20) 
ax.set_ylabel(r'$\theta_1$',fontsize=20)
ax.set_zlabel('J(θ)')
ax.scatter(xvals, yvals, zvals,c=np.abs(zvals),cmap=plt.cm.hot)#生成一个曲面
plt.title('Cost and optimization path',fontsize=20)
plt.plot([theta[0] for theta in thetaHistory],[theta[1] for theta in thetaHistory],costHistory,'bo-')
plt.show()
```

### Linear regression with multiple variables
In this part, you will implement linear regression with multiple variables to predict the prices of houses. Suppose you are selling your house and you want to know what a good market price would be. One way to do this is to first collect information on recent houses sold and make a model of housing prices.

The file ex1data2.txt contains a training set of housing prices in Portland, Oregon. The first column is the size of the house (in square feet), the second column is the number of bedrooms, and the third column is the price of the house.

```python
data = pd.read_csv('ex1/ex1data2.txt',header=None)
data.head(10)
```

Once the former code has support multivariable linear regression,we can dircetly use it.We ecapsulate it first.

```python
#Linear Regression
#traning set is nxm matrix,no need to manual add bias column(x0=1)

class LinearRegression:
    def __init__(self,trainingSet,labels,iterations=1500,alpha=0.01):
        self.alpha = alpha
        self.iterations = iterations
        self.y = labels
        self.m = trainingSet.shape[0] #no. of examples
        self.n = trainingSet.shape[1]+1 #no. of feature
        self.costHistory = []
        self.thetaHistory = []
        assert(len(labels)==self.m)
        self.theta = np.zeros([self.n,1]) #init parameters
        self.X = np.c_[np.ones(self.m),trainingSet] #add a column of ones to x
        #print(self.X.shape,self.y.shape,self.theta.shape)
    
    #if n is large,use gradient discent,otherwise use normal equation
    def fit(self):
        if(self.n<1000):
            print('using normal equation...')
            self.theta = self.__normalEquation(self.X,self.y)
        else:
            print('using gradient descent...')
            self.theta, self.thetaHistory, self.costHistory = self.__gradientDescent(self.X,self.y,self.theta,self.alpha,self.iterations)
        return self.theta, thetaHistory, self.costHistory
    
    #use gradientDescent
    def gradientDescent(self):
        self.theta, self.thetaHistory, self.costHistory = self.__gradientDescent(self.X,self.y,self.theta,self.alpha,self.iterations)
        return self.theta, thetaHistory, self.costHistory
    
    #use normalEquation
    def normalEquation(self):
        self.theta = self.__normalEquation(self.X,self.y)
        return self.theta
    
    #make hypothesis    
    def predict(self,X):
        return self.__h(self.theta,X)
    
    #plot learning curve
    def plotConvergence(self):
        costHistory = self.costHistory
        iterations = self.iterations
        plt.figure(figsize=(10,6))
        plt.grid(True)
        plt.title("Convergence of Cost Function")
        plt.xlabel("Iteration number")
        plt.ylabel("Cost function")
        plt.xlim([-0.1*iterations,1.1*iterations])
        plt.ylim([0.1*np.min(costHistory),1.1*np.max(costHistory)])
        plt.plot(range(iterations),costHistory,'bo')
        
    #hypothesis function 
    def __h(self,theta,X):
        return np.dot(X,theta)

    #computing the cost J(θ)
    def __J(self,theta,X,y):
        m = self.m
        cost = 0.5/m*np.dot((self.__h(theta,X)-y).T,self.__h(theta,X)-y)
        return cost[0][0]

    #gradient descent
    def __gradientDescent(self,X,y,theta,alpha,iterations):
        m = self.m
        thetaHistory = [theta] #for plotting theta-iter
        costHistory = [self.__J(theta,X,y)] #for plotting J-iter
        #start iteration
        for iter in range(1,iterations):
            theta = theta - alpha*1/m*np.dot(X.T,self.__h(theta,X)-y)
            thetaHistory.append(theta)
            costHistory.append(J(theta,X,y))
        return theta, thetaHistory, costHistory

    #normal equation
    def __normalEquation(self,X,y):
        return np.dot(np.dot(np.linalg.inv(np.dot(X.T,X)),X.T),y)
```

```python
trainingSet = data[[0,1]]#no need add a column of ones to x
labels = data[[2]]
trainingSet.head(5)
```

### Standardization

```python
#X is dataset matrix,each row is an example
def standardization(X):
    mu = np.mean(X,axis = 0)
    sigma = np.std(X,axis = 0)
    return (X - mu)/sigma

trainingSet = standardization(trainingSet)
trainingSet.head(5)
```

```python
#make a linear regression instance
lr = LinearRegression(trainingSet,labels,iterations=1500,alpha=0.1)
#theta by gradient descent
theta, thetaHistory, costHistory = lr.gradientDescent()
theta
```

```python
#plot learning curve
lr.plotConvergence()
```

```python
#theta by normal equation
theta = lr.normalEquation()
theta
```
