###  1.3 Vectorizing Logistic Regression (Regularized)

```python
#hypotheis function for logistic regression
def h(theta,X):
    return sigmod(np.dot(X,theta))
```

#### 1.3.1 Vectorizing the cost function

```python
#cost function for logistic regression
def J(theta,X,y,lamda=0):
    m = X.shape[0]
    term1 = np.dot(-np.array(y).T,np.log(h(theta,X)))
    term2 = np.dot((1-np.array(y)).T,np.log(1-h(theta,X)))
    regTerm = (lamda/2.0)*np.sum(np.dot(theta[1:].T,theta[1:])) #Skip theta0
    return 1.0/m*(np.sum(term1 - term2) + regTerm)
```

#### 1.3.2 Vectorizing the gradient

```python
#instead use fminunc function, use gradient descent again
#gradient descent for logistic regression 
def costGradient(theta,X,y,lamda=0):
    m = X.shape[0]
    return 1/m*np.dot(X.T,h(theta,X)-y) + lamda/m*theta #grad vector
```

### 1.4 One-vs-all Classification

```python
from scipy import optimize

def optimizeTheta(theta,X,y,lamda=0.):
    result = optimize.fmin_cg(J, fprime=costGradient, x0=theta, args=(X, y, lamda), maxiter=80, disp=False, full_output=True)
    return result[0], result[1]
```

```python
Theta = np.zeros((10,X.shape[1]))
lamda = 0
for i in range(10):
    print('optimize for number %s...' %i )
    class_i = i if i else 10
    logic_Y = np.array([1 if x == class_i else 0 for x in y])
    theta_i = np.zeros((X.shape[1],1)).reshape(-1)
    theta_i,mincost = optimizeTheta(theta_i,X,logic_Y,0)
    Theta[i,:] = theta_i
print('Done.')
```

#### 1.4.1 One-vs-all Prediction

```python
#predict result is negative or positive
def predict(theta,X):
    return h(sigmod(np.dot(X,theta)))>=0.5

#one-vs-all classification
def predictOneVsAll(Theta,X):
    ans = np.argmax(sigmod(np.dot(Theta,X)))
    return ans if ans else 10 
```
