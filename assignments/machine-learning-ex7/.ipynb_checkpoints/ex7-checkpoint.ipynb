#### 1.1 Find closest centroids

```python
#X is (m,n) darray, centriod is (K,n) darray, return a  (m,) darray of index of K points  
def findClosestCentroids(X, centroids):
    m = X.shape[0]
    idx = []
    for i in range(m):
        D = centroids - X[i, :] #(K,n) 
        D = np.sum(D**2, axis = 1) #(K,)
        idx.append(np.argmin(D))
    return idx
```

```python
idx = findClosestCentroids(X,initial_centroids)
idx[:3] #select the 0,1,2 points, find the closest points, return the 1st, 3rd, 2
```

```python
#centroidsHistory is an list, centroids[i] is a (K,n) darray
def plotKClusters(X, idx, centroidsHistory):
    assert len(centroidsHistory)>0
    K = len(centroidsHistory[0])
    colors =  ['b','g','gold','darkorange','salmon','olivedrab']
    m = X.shape[0]
    kXset = []
    plt.figure(figsize=(10,7))
    plt.grid(ls='--')
    plt.xlabel('${x_0}$', fontsize = 20)
    plt.ylabel('${x_1}$', fontsize = 20)
    plt.title('Plot %d Clusters' % K ,fontsize = 20)
    for i in range(K):
        kXset.append([])
        x = [centriods[:, 0] for centriods in centroidsHistory]
        y = [centriods[:, 1] for centriods in centroidsHistory]
        plt.plot(x, y,'rx--',markersize=10)
    for i in range(m):
        kXset[idx[i]].append(X[i, :])
    for k in range(K):
        Xset = kXset[k]
        color = colors[k]
        xvals = [x[0] for i,x in enumerate(Xset)]
        yvals = [x[1] for i,x in enumerate(Xset)]
        plt.scatter(xvals, yvals, color=color, label='Cluster %d' %k)
    plt.legend()
```

```python
plotKClusters(X,idx,[initial_centroids])
```

#### 1.1.2 Computing centroid means

```python
#X is (m,n) darray, centroids is (K,n) darray, idx is (m,) array, K is scalar number, return a (K,n) darray
def computeMeans(X, idx, K):
    n = X.shape[1] #dimensions
    m = X.shape[0] #no. of X
    mu = np.zeros([K, n])
    for k in range(K):
        p = [ i for i,x in enumerate(idx) if x==k ] #idx of points that are assigned to k th cluster
        mu[k , :] = np.sum(X[p], axis = 0).reshape([1,n])/len(p)
    return mu
```

#### KMeans Wrap

```python
class KMeans:
    #__() functions are private pure function
    
    #X is (m,n) darray, with m examples, each example has n dims
    def __init__(self,X):
        self.X = X
        
    #kmeans cost function
    #X is (m,n) darray, centroids is (K,n) darray, idx is (m,) array, return a scalar number
    def __J(self, X, centroids, idx):
        m = X.shape[0]
        m = X.shape[1]
        K = centroids.shape[0]
        cost = 0.
        for i in range(m):
            d = X[i] - centroids[idx[i]] #(1,n)
            cost += np.dot(d, d.T)
        return cost/m
    
    #X is a (m,n) darray, random select K point ,return a  (K,n) darray
    def __kMeansInitCentroids(self, X, K):
        m  = X.shape[0] #no. of data
        centroids = X[random.sample(range(m), K)]
        return centroids
    
    #X is (m,n) darray, centriod is (K,n) darray, return a  (m,) darray of index of K points  
    def __findClosestCentroids(self, X, centroids):
        m = X.shape[0]
        idx = []
        for i in range(m):
            D = centroids - X[i, :] #(K,n) 
            D = np.sum(D**2, axis = 1) #(K,)
            idx.append(np.argmin(D))
        return idx
    
    #X is (m,n) darray, centroids is (K,n) darray, idx is (m,) array, K is scalar number, return a (K,n) darray
    def __computeMeans(self, X, idx, K):
        n = X.shape[1] #dimensions
        m = X.shape[0] #no. of X
        mu = np.zeros([K, n])
        for k in range(K):
            p = [ i for i,x in enumerate(idx) if x==k ] #idx of points that are assigned to k th cluster
            mu[k , :] = np.sum(X[p], axis = 0).reshape([1,n])/len(p)
        return mu
    
    # random init and return idx and cost, may converge to local optima
    # X is (m,n) darray, K is scalar number
    def __randomKMeans(self, X, K):
        centroids = self.__kMeansInitCentroids(X, K) #init K centroids
        idx = self.__findClosestCentroids(X, centroids) #init idx
        cost = self.__J(X, centroids, idx) #init cost
        maxIterations = 100 #prevent dead loop
        iterIdx = 0
        while(iterIdx<maxIterations):
            iterIdx = iterIdx + 1
            idx = self.__findClosestCentroids(X, centroids)
            centroids = self.__computeMeans(X, idx, K)
            newCost = self.__J(X, centroids, idx)
            if(newCost==cost): #J decrease everytime, if it remains unchange its converged
                break
            cost = newCost
        return idx, cost, centroids
    
    # multi-times random init and mat converge to global optima
    # X is (m,n) darray, K is scalar number
    def findKCluster(self, K, maxIterations = 20, printLog = True):
        X = self.X
        costHistory = [] 
        idxHistory = []
        centroidsHistory = []
        for i in range(maxIterations):
            idx, cost, centroids = self.__randomKMeans(X, K)
            if(printLog):
                print('iteration %d , cost = %f' %(i+1, cost))
            costHistory.append(cost)
            idxHistory.append(idx)
            centroidsHistory.append(centroids)
        optimaIter = np.argmin(costHistory)
        costOptima = costHistory[optimaIter]
        idxOptima = idxHistory[optimaIter]
        centroidOptima = centroidsHistory[optimaIter]
        return costOptima, idxOptima, centroidOptima
```

```python
kmeans = KMeans(X)
cost, idx, centroids = kmeans.findKCluster(3,10)
plotKClusters(X,idx,[centroids])
```

### 1.4 Image compression with K-means
In this exercise, you will apply K-means to image compression. In a straightforward 24-bit color representation of an image,each pixel is represented as three 8-bit unsigned integers (ranging from 0 to 255) that specify the red, green and blue intensity values. This encoding is often refered to as the RGB encoding. Our image contains thousands of colors, and in this part of the exercise, you will reduce the number of colors to 16 colors.

By making this reduction, it is possible to represent (compress) the photo in an efficient way. Specifically, you only need to store the RGB values of the 16 selected colors, and for each pixel in the image you now need to only store the index of the color at that location (where only 4 bits are necessary to represent 16 possibilities).

In this exercise, you will use the K-means algorithm to select the 16 colors that will be used to represent the compressed image. Concretely, you will treat every pixel in the original image as a data example and use the K-means algorithm to find the 16 colors that best group (cluster) the pixels in the 3-dimensional RGB space. Once you have computed the cluster centroids on the image, you will then use the 16 colors to replace the pixels in the original image.

```python
imgMtx = np.asarray(img)
imgMtx.shape
```

```python
imgVector = imgMtx.reshape(-1,3)/255.
imgVector.shape
```

```python
K = 16 # 24bit(RGB) to 16 colors
kmeans = KMeans(imgVector)
cost, idx, centroids = kmeans.findKCluster(K,maxIterations = 10)
```

## 2 Principal Component Analysis
In this exercise, you will use principal component analysis (PCA) to perform dimensionality reduction. You will first experiment with an example 2D dataset to get intuition on how PCA works, and then use it on a bigger dataset of 5000 face image dataset.

### 2.1 Example Dataset

```python
datafile = 'ex7/ex7data1.mat'
mat = scipy.io.loadmat( datafile )
X = mat['X']
print('X shape = %s' %(X.shape,))
```

```python
#plot data
plt.figure(figsize=(9,9))
plot = plt.scatter(X[:,0], X[:,1], s=30, facecolors='none', edgecolors='b')
plt.title("Example Dataset",fontsize=18)
plt.grid('--')
```

### 2.2 Implementing PCA

```python
#standard normalization
#X is (m,n) darray return (n,) (n,) (m,n) darray
def featureNormalize(X):
    means = np.mean(X, axis = 0)
    Xnorm = X - means
    stds = np.std(Xnorm, axis = 0)
    Xnorm = Xnorm / stds
    return means, stds, Xnorm
```

```python
#start PCA
#step.1 feature normalization
means, stds, Xnorm = featureNormalize(X)
#step.2 calculate covariance matrix
m = Xnorm.shape[0]
covMatrix = 1./m*np.dot(Xnorm.T, Xnorm) 
#step.3 run single value decomposition to get U,S,V
U, S, V = scipy.linalg.svd(covMatrix, full_matrices = True, compute_uv = True)
```

```python
# "...output the top principal component (eigen- vector) found, 
# and you should expect to see an output of about [-0.707 -0.707]"
print('Top principal component is %s'%U[:,0])
```

```python
#Quick plot, now including the principal component
plt.figure(figsize=(9,9))
plot = plt.scatter(X[:,0], X[:,1], s=30, facecolors='none', edgecolors='b')
plt.title("Example Dataset: PCA Eigenvectors Shown",fontsize=18)
plt.xlabel('${x_1}$',fontsize=18)
plt.ylabel('${x_2}$',fontsize=18)
plt.grid('--')
#To draw the principal component, you draw them starting
#at the mean of the data
plt.plot([means[0], means[0] + 1.5*S[0]*U[0,0]], 
         [means[1], means[1] + 1.5*S[0]*U[0,1]],
        color='red',linewidth=3,
        label='First Principal Component')
plt.plot([means[0], means[0] + 1.5*S[1]*U[1,0]], 
         [means[1], means[1] + 1.5*S[1]*U[1,1]],
        color='fuchsia',linewidth=3,
        label='Second Principal Component')

leg = plt.legend(loc=4)
```

### 2.3 Dimensionality Reduction with PCA

#### 2.3.1 Projecting the data onto the principal components

```python
#project (m,n) shaped X to (M,k) shaped data 
def projectData(X, U, K):
    Ureduced = U[:, :K]
    Z = np.dot(X, Ureduced)
    return Z
```

```python
Z = projectData(Xnorm, U, 1)
print('X shape = %s , Z shape = %s' %(Xnorm.shape,Z.shape,))
print('X[0] = %s , Z[0] = %s' %(X[0],Z[0]))
```

#### 2.3.2 Reconstructing an approximation of the data

```python
def reconstructData(Z, U, K):
    Ureduced = U[:,:K]
    Xapprox = np.dot(Z, Ureduced.T)
    return Xapprox
```

```python
Xrecovered = reconstructData(Z, U, 1)
print('recovered approximation of X[0] is %s' %Xrecovered[0])
```

#### 2.3.3 Visualizing the projections

```python
plt.figure(figsize=(9,9))
plot = plt.scatter(Xnorm[:,0], Xnorm[:,1], s=30, facecolors='none', 
                   edgecolors='b',label='Original Data Points')
plot = plt.scatter(Xrecovered[:,0], Xrecovered[:,1], s=30, facecolors='none', 
                   edgecolors='r',label='PCA Reduced Data Points')

plt.title("Example Dataset: Reduced Dimension Points Shown",fontsize=14)
plt.xlabel('x1 [Feature Normalized]',fontsize=14)
plt.ylabel('x2 [Feature Normalized]',fontsize=14)
plt.grid('--')

for x in range(Xnorm.shape[0]):
    plt.plot([Xnorm[x,0],Xrecovered[x,0]],[Xnorm[x,1],Xrecovered[x,1]],'k--')

leg = plt.legend(loc=4)
```

### 2.4 Face Image Dataset

```python
datafile = 'ex7/ex7faces.mat'
mat = scipy.io.loadmat(datafile)
X = mat['X']
print('face img array X shape = %s' %(X.shape,))
```

```python
def scalePixels(X):
    #make it at the range of 0-255,or picture will be too dark
    minv = np.min(X)
    maxv = np.max(X)
    newX = ((X - minv)/(maxv-minv)*255).astype(np.uint8)
    return newX

def ImageRowToMatrix(row,width = 32, height = 32):
    mtx = row.reshape(width,height)
    return scalePixels(mtx.T)
    
def displayBigPicture(X, nrows = 10, ncols = 10,width = 32, height =  32):
    """
    Function that picks the first 100 rows from X, creates an image from each,
    then stitches them together into a 10x10 grid of images, and shows it.
    """
    big_picture = np.zeros((height*nrows,width*ncols))
    
    irow, icol = 0, 0
    for idx in range(nrows*ncols):
        if icol == ncols:
            irow += 1
            icol = 0
        imgMtx = ImageRowToMatrix(X[idx])
        big_picture[irow*height:irow*height+imgMtx.shape[0],icol*width:icol*width+imgMtx.shape[1]] = imgMtx
        icol += 1
        
    fig = plt.figure(figsize=(10,10))
    img = Image.fromarray(big_picture)
    plt.imshow(img,cmap = cm.Greys)
    
displayBigPicture(X)
```

#### 2.4.1 PCA on Faces

```python
#start PCA
#step.1 feature normalization
means, stds, Xnorm = featureNormalize(X)
#step.2 calculate covariance matrix
m = Xnorm.shape[0]
covMatrix = 1./m*np.dot(Xnorm.T, Xnorm) 
#step.3 run single value decomposition to get U,S,V
U, S, V = scipy.linalg.svd(covMatrix, full_matrices = True, compute_uv = True)
```

```python
displayBigPicture((U[:,:36]).T, nrows = 6, ncols = 6)
```

#### 2.4.2 Dimensionality Reduction

```python
# Project each image down to 36 dimensions
Z = projectData(Xnorm, U, K = 36)
# Attempt to recover the original data
Xrecover = reconstructData(Z, U, K = 36)
displayBigPicture(Xrecover)
```
