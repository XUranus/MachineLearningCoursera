```python
#plot the data
def plotData(pos,neg,posLabel,negLabel,xlabel,ylabel):
    plt.scatter(pos[0],pos[1],c='black',marker='+',label=posLabel)
    plt.scatter(neg[0],neg[1],c='orange',marker='o',label=negLabel)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend()

plt.figure(figsize=(10,7))
plotData(pos,neg,'Admitted','Not Admitted','Exam 1 score','Exam 2 score')
```

### 1.2 Implementation

#### 1.2.1 sigmod function

```python
#vectorized sigmod function(perform element wise on matrix)
def sigmod(x):
    return 1/(1+np.exp(-1*x))

plt.figure(figsize=(10,6))
plt.plot(np.arange(-5,5,0.1),sigmod(np.arange(-5,5,0.1)))
plt.title('sigmod function')
plt.grid(True)
```

#### 1.2.2 Cost function and gradient

```python
#hypotheis function for logistic regression
def h(theta,X):
    return sigmod(np.dot(X,theta))

#cost function for logistic regression
def J(theta,X,y,lamda=0):
    m = X.shape[0]
    term1 = np.dot(-np.array(y).T,np.log(h(theta,X)))
    term2 = np.dot((1-np.array(y)).T,np.log(1-h(theta,X)))
    regTerm = (lamda/2.0)*np.sum(np.dot(theta[1:].T,theta[1:])) #Skip theta0
    return 1.0/m*(np.sum(term1 - term2) + regTerm)

#predict result is negative or positive
def predict(theta,X):
    return h(sigmod(np.dot(X,theta)))>=0.5

#instead use fminunc function, use gradient descent again
#gradient descent for logistic regression 
def gradientDescent(X,y,theta,alpha,iterations,lamda=0):
    m = X.shape[0]
    thetaHistory = [theta] #for plotting theta-iter
    costHistory = [J(theta,X,y,lamda)] #for plotting J-iter
    #start iteration
    for iter in range(1,iterations):
        grad = 1/m*np.dot(X.T,h(theta,X)-y) + lamda/m*theta #grad vector
        theta = theta - alpha*grad
        thetaHistory.append(theta)
        costHistory.append(J(theta,X,y,lamda))
    return theta, thetaHistory, costHistory
```

```python
#prepare for the data X,y
X = np.c_[np.ones(data.shape[0]),data[[0,1]]] #add a column of ones to x
y = np.reshape(data[2].values,(len(data[2]),1),order='C')
theta = np.zeros((X.shape[1],1)) #initialize theta

#compute current cost,parameter lamda temporarily set as 0
print('current cost = %s' %J(theta,X,y))
```

#### 1.2.3 Learning parameters using fminunc

```python
#An alternative to octave's "fminunc"
#http://docs.scipy.org/doc/scipy-0.16.0/reference/generated/scipy.optimize.fmin.html
from scipy import optimize

def optimizeTheta(theta,X,y,lamda=0.):
    result = optimize.fmin(J, x0=theta, args=(X, y, lamda), maxiter=500, full_output=True)
    return result[0], result[1] #theta,mincost

theta,cost = optimizeTheta(theta,X,y)
theta = np.reshape(theta,(3,1),order='C')
print('theta = %s cost = %s' %(theta,cost))
```

```python
#if still using gradient descent
theta2 = np.zeros((X.shape[1],1)) #initialize theta
theta2,thetaHistory,costHistory = gradientDescent(X,y,theta2,0.001,10000000,0)
print('theta = %s cost = %s' %(theta2,J(theta2,X,y)))
```

```python
def plotConvergence(costHistory):
    iterations = len(costHistory)
    plt.figure(figsize=(10,6))
    plt.grid(True)
    plt.title("Convergence of Cost Function")
    plt.xlabel("Iteration number")
    plt.ylabel("Cost function")
    plt.xlim([-0.1*iterations,1.1*iterations])
    plt.plot(range(len(costHistory)),costHistory,'bo')

plotConvergence(costHistory)
#in this case gradient descent really slow, 
#for learning rate > 0.001 won't converge
```

```python
#plot the decision boundary
def plotDecisonBoundary(theta,X):    
    boundary_xs = np.array([np.min(X[:,1]), np.max(X[:,1])])
    boundary_ys = (-1./theta[2])*(theta[0] + theta[1]*boundary_xs)
    plotData(pos,neg,'Admitted','Not Admitted','Exam 1 score','Exam 2 score')
    plt.grid(True)
    plt.plot(boundary_xs,boundary_ys,'b-',label='Decision Boundary')
    plt.legend()

plotDecisonBoundary(theta,X)
```

### 2 Regularized logistic regression
In this part of the exercise, you will implement regularized logistic regression to predict whether microchips from a fabrication plant passes quality assurance (QA). During QA, each microchip goes through various tests to ensure it is functioning correctly.

Suppose you are the product manager of the factory and you have the
test results for some microchips on two different tests. From these two tests, you would like to determine whether the microchips should be accepted or rejected. To help you make the decision, you have a dataset of test results on past microchips, from which you can build a logistic regression model.

```python
data = pd.read_csv('ex2/ex2data2.txt',header=None)
pos = data[data[2]==1] #positive example2
neg = data[data[2]==0] #negative examples
data.head(5)
```

#### 2.1 Visualizing the data

```python
plotData(pos,neg,'y=1','y=0','microchip test 1','microchip test 2')
```

#### 2.2 Feature mapping
One way to fit the data better is to create more features from each data point.

```python
# copied
def mapFeature( x1col, x2col ):
    """ 
    Function that takes in a column of n- x1's, a column of n- x2s, and builds
    a n- x 28-dim matrix of featuers as described in the homework assignment
    """
    degrees = 6
    out = np.ones( (x1col.shape[0], 1) )

    for i in range(1, degrees+1):
        for j in range(0, i+1):
            term1 = x1col ** (i-j)
            term2 = x2col ** (j)
            term  = (term1 * term2).reshape( term1.shape[0], 1 ) 
            out   = np.hstack(( out, term ))
    return out
```

```python
#prepare for the data X,y
X = np.c_[np.ones(data.shape[0]),data[[0,1]]] #add a column of ones to x
y = np.reshape(data[2].values,(len(data[2]),1),order='C')
```

```python
mappedX = mapFeature(X[:,1],X[:,2])
theta = np.zeros((mappedX.shape[1],1)) #initialize theta , theta dimension should correspond with mapped X
print('X shape = %s    =>    mappedX shape = %s' %(X.shape,mappedX.shape))
```

#### 2.3 Cost function and gradient

```python
J(theta,mappedX,y)
```

#### 2.3.1 Learning parameters using fminunc

```python
theta,cost = optimizeTheta(theta,mappedX,y)
theta = np.reshape(theta,(mappedX.shape[1],1),order='C')
print('optima cost = %s' %(cost))
```

```python
#notice using 'fmin' cost did not converge, using minimize instead
def optimizeRegularizedTheta(theta,X,y,lamda=0.):
    result = optimize.minimize(J, theta, args=(X, y, lamda),  method='BFGS', options={"maxiter":500, "disp":False} )
    theta, cost = np.array([result.x]), result.fun
    theta = np.reshape(theta,(X.shape[1],1),order='C')
    return theta,cost

theta = np.zeros((mappedX.shape[1],1)) #initialize theta 
theta, cost = optimizeRegularizedTheta(theta,mappedX,y)
print('cost = %s' %(cost))
```

#### 2.4 Plotting the decision boundary

```python
#copied
def plotBoundary(mytheta, myX, myy, mylambda=0.):
    """
    Function to plot the decision boundary for arbitrary theta, X, y, lambda value
    Inside of this function is feature mapping, and the minimization routine.
    It works by making a grid of x1 ("xvals") and x2 ("yvals") points,
    And for each, computing whether the hypothesis classifies that point as
    True or False. Then, a contour is drawn with a built-in pyplot function.
    """
    theta, mincost = optimizeRegularizedTheta(mytheta,myX,myy,mylambda)
    xvals = np.linspace(-1,1.5,50)
    yvals = np.linspace(-1,1.5,50)
    zvals = np.zeros((len(xvals),len(yvals)))
    for i in range(len(xvals)):
        for j in range(len(yvals)):
            myfeaturesij = mapFeature(np.array([xvals[i]]),np.array([yvals[j]]))
            zvals[i][j] = np.dot(theta.T,myfeaturesij.T)
    zvals = zvals.transpose()

    u, v = np.meshgrid( xvals, yvals )
    mycontour = plt.contour( xvals, yvals, zvals, [0])
    #Kind of a hacky way to display a text on top of the decision boundary
    myfmt = { 0:'Lambda = %d'%mylambda}
    plt.clabel(mycontour, inline=1, fontsize=15, fmt=myfmt)
    plt.title("Decision Boundary")
```

```python
plt.figure(figsize=(12,10))
plt.subplot(221)
plotData(pos,neg,'y=1','y=0','microchip test 1','microchip test 2')
plotBoundary(theta,mappedX,y,0.)

plt.subplot(222)
plotData(pos,neg,'y=1','y=0','microchip test 1','microchip test 2')
plotBoundary(theta,mappedX,y,1.)

plt.subplot(223)
plotData(pos,neg,'y=1','y=0','microchip test 1','microchip test 2')
plotBoundary(theta,mappedX,y,10.)

plt.subplot(224)
plotData(pos,neg,'y=1','y=0','microchip test 1','microchip test 2')
plotBoundary(theta,mappedX,y,100.)
```
